# 14. Какая команда выиграла наибольшое количество матчей в сезоне? (Будьте внимательны с типом матча в сезоне - season).

Чтобы определить команду (или команды), которая выиграла наибольшее количество матчей в сезоне, мы создаем датафрейм с названием команды и ее результатом игры.
Для этого соединяем название гостевой команды 'away_team' и результат матча для нее 'home_team_loss' с названием домашней команды 'home_team' и ее результата 'home_team_win'. Производим переименование колонок для удобства и получаем датафрейм с названием команды и результатом матча для нее. Затем группируем данные по названию команды и с помощью sum() подсчитываем количество побед. 

Из полученного датафрейма выводим все строки с максимальным значением количества выигрышей ('win') и выводим значение столбца 'team' 

```python
team_win_count = pd.concat([data[['away_team', 'home_team_loss']].rename(columns={'away_team':'team', 'home_team_loss':'win'}), data[['home_team', 'home_team_win']].rename(columns={'home_team':'team', 'home_team_win':'win'})]).groupby(['team']).sum('win').reset_index('team')
team_win_count[team_win_count['win'] == team_win_count['win'].max()]['team']
```

# 15. Какая команда выиграла наибольшее количество домашних матчей в сезоне?

Чтобы определить команду (или команды), которая выиграла наибольшее количество домашних матчей в сезоне, мы создаем датафрейм с суммой выигранных матчей (sum('home_team_win')) для каждой команды groupby('home_team'). 

Из полученного датафрейма выводим все строки с максимальным значением количества выигрышей ('home_team_win') и выводим значение столбца 'team' 

```python
home_team_wins = data[['home_team','home_team_win']].groupby('home_team').sum('home_team_win')
home_team_wins[home_team_wins['home_team_win'] == home_team_wins['home_team_win'].max()].index[0]
```

# 18. Зависит ли выигрыш от количества посетителей матча? (https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.cov.html)

Допустим, что имеется в виду выигрыш домашней команды и исследуется его зависимость от количества зрителей. 
Построим ковариационную матрицу по этим столбцам

Однозначных выводов по кавариации сделать невозможно, так как она зависит от дисперсий распределения количества зрителей и распределения выигрышей.
